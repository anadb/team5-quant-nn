```python
import yaml

with open("/home/team5/examples/output_task1/solution.yaml", 'rt', encoding='UTF-8') as f:
  solutions = yaml.safe_load(f.read())
```

```python
def decode(x):
  # (n - 1) * 96 + offset
  return (x // 96 + 1, x % 96 + 1)
```

```python
decode(1 - 1)
```

```python
ANS_A = []
ANS_B = []

for k, v in solutions[0]["solution"]["variables"].items():
  x = int(k.split("_")[1])
  if x <= 960:
    row, col = decode(x - 1)
    if int(v):
      ANS_A.append((1, row, col))
    else:
      ANS_A.append((0, row, col))
    # ANS_A[row - 1][col - 1] = int(v)
  elif x <= 1920:
    row, col = decode(x - 960 - 1)
    if int(v):
      ANS_B.append((1, row, col))
    else:
      ANS_B.append((0, row, col))
    # ANS_B[row - 1][col - 1] = int(v)
  
```
