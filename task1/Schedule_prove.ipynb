```python
custom_dict_subject = {'Квантовая механика': 0, 'Квантовая теория информации': 1, 'Квантовые вычисления': 2, 'Сложность квантовых алгоритмов': 3, 'Квантовые алгоритмы в логистике': 4,
                        'Квантовое машинное обучение': 5, 'Моделирование квантовых систем': 6, 'Квантовые алгоритмы в химии': 7,'Физическая реализация квантовых компьютеров': 8,
                        'Моделирование квантовых алгоритмов': 9, 'нет занятий': 10}
custom_dict_subjectyn = {'Квантовая механика': 1, 'Квантовая теория информации': 1, 'Квантовые вычисления': 1, 'Сложность квантовых алгоритмов': 1, 'Квантовые алгоритмы в логистике': 1,
                        'Квантовое машинное обучение': 1, 'Моделирование квантовых систем': 1, 'Квантовые алгоритмы в химии': 1,'Физическая реализация квантовых компьютеров': 1,
                        'Моделирование квантовых алгоритмов': 1, 'нет занятий': 0}

custom_dict_teacher = {'Иванов': 0, 'Петров': 1, 'Сидоров': 2, 'Карпов': 3, 'Соколов': 4, 'нет занятий': 10}
custom_dict_teacheryn = {'Иванов': 1, 'Петров': 1, 'Сидоров': 1, 'Карпов': 1, 'Соколов': 1, 'нет занятий': 0}


pll['Subject_N'] =  pll['Subject'].apply(lambda x: custom_dict_subject[x])
pll['Subject_yn'] =  pll['Subject'].apply(lambda x: custom_dict_subjectyn[x])
pll['Teacher_N'] =  pll['Name'].apply(lambda x: custom_dict_teacher[x])
pll['Teacher_yn'] =  pll['Name'].apply(lambda x: custom_dict_teacheryn [x])


pll['Day_N_thru'] = pll['Week_N']*6 + pll['Day_N']

pll['Hour_N_thru'] = pll['Group_N']*2*6*8 + pll['Week_N']*6*8 + pll['Day_N']*8 + pll['Hour_N']

pll
```

```python
#0.1.1 У двух групп занятия не могут идти одновременно с одним преподавателем
import numpy as np

def prove_011():
    pl1 = pll[(pll.Group =='QC_1')].reset_index(drop=True)
    pl2 = pll[(pll.Group =='QC_2')].reset_index(drop=True)
    plcompare = pl1

    plcompare['equal'] = ((pl1['Subject'] == pl2['Subject']) & (pl1['Subject'] != 'нет занятий'))

    return bool(not (plcompare['equal']=='False').sum())

prove_011()
```

```python
#0.2 Преподаватель не может одновременно вести два предмета

def prove_02():
    pl1 = pl[(pl.Group =='QC_1') ].reset_index(drop=True)
    pl2 = pl[(pl.Group =='QC_2') ].reset_index(drop=True)
    plcompare = pl1
    plcompare['equal'] = (pl1['Name'] == pl2['Name']) & (pl2['Name'] != 'нет занятий')

    return bool(not (plcompare['equal']=='False').sum())

prove_02()
```

```python
#0.3 У группы может идти только одно занятие в срезе группа-день-час

def prove_03():
    pl1 = pl[(pl.Group =='QC_1') & (pl.Subject != 'нет занятий')].reset_index(drop=True)
    pl2 = pl[(pl.Group =='QC_2') & (pl.Subject != 'нет занятий')].reset_index(drop=True)
    s1 =len(pl1['Hour_N_thru'])- len(pl1['Hour_N_thru'].drop_duplicates())
    s2= len(pl2['Hour_N_thru'])- len(pl2['Hour_N_thru'].drop_duplicates())
    return bool(not (s1+s2)),s1,s2

prove_03()

```

```python
#2. В день может быть поставлено не более 6 лекций в расписании группы

def prove_02():
    pl1 = pl[(pl.Group =='QC_1') ].reset_index(drop=True)
    pl2 = pl[(pl.Group =='QC_2') ].reset_index(drop=True)

    #lessons_per_day_gr1 = []
    #for i in range(len(pl1)):
    s1 =pl1.groupby(['Day_N_thru'])['Subject_yn'].sum()
    s2 =pl2.groupby(['Day_N_thru'])['Subject_yn'].sum()

    return bool((s1.max()<=6) & (s2.max() <=6)), s1,s2

prove_02()
```

```python
#3. В день не может быть более двух занятий по одному предмету у группы.

def prove_03():
    pl1 = pl[(pl.Group =='QC_1') ].reset_index(drop=True)
    pl2 = pl[(pl.Group =='QC_2') ].reset_index(drop=True)

    #lessons_per_day_gr1 = []
    #for i in range(len(pl1)):
    s1 =pl1.groupby(['Day_N_thru', 'Subject'])['Subject_yn'].sum()
    s2 =pl2.groupby(['Day_N_thru', 'Subject'])['Subject_yn'].sum()

    return bool((s1.max()<=2) & (s2.max() <=2)), s1,s2

prove_03()
```

```python
#4 Так получилось, что преподаватель n1 не может вести лекции в среду,
#а n2 в понедельник, а n3 никак не может работать по субботам, n4, n5 –
#не могут вести занятия по вторникам. Воскресенье - день без занятий

def prove_04():

  n1 = pl[(pl['Name']=='Иванов') & (pl['Day']=='cр')].groupby(['Subject'])['Subject_yn'].count()
  n2 = pl[(pl['Name']=='Петров') &(pl['Day']=='пн')].groupby(['Subject'])['Subject_yn'].count()
  n3 = pl[(pl['Name']=='Сидоров') &(pl['Day']=='сб')].groupby(['Subject'])['Subject_yn'].count()
  n4 = pl[(pl['Name']=='Карпов') &(pl['Day']=='вт')].groupby(['Subject'])['Subject_yn'].count()
  n5 = pl[(pl['Name']=='Соколов') &(pl['Day']=='вт')].groupby(['Subject'])['Subject_yn'].count()


  return  bool((len(n1)== 0) & (len(n2)== 0) & (len(n3)== 0) & (len(n4)== 0) & (len(n5)== 0)), len(n1),len(n2),len(n3),len(n4),len(n5)

prove_04()
```

```python
# 5.По требованию учебной программы для каждого предмета должно быть выделено
# 4 часа за период T = 2 недели для каждой группы.

def prove_05():

    s1 =pl[(pl.Group =='QC_1') & (pl.Subject != 'нет занятий')].groupby(['Subject'])['Subject_yn'].sum()
    s2 =pl[(pl.Group =='QC_2') & (pl.Subject != 'нет занятий')].groupby(['Subject'])['Subject_yn'].sum()

    return bool((s2.min() == 4) & (s2.max() == 4) & (s2.min() == 4) & (s2.max() == 4)), s1, s2

prove_05()
```
